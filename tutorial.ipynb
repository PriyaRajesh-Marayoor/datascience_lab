```python
import numpy as np
arr=np.array([[2,2,3],
             [4,1,2],
             [5,7,8]])
print(arr)
t=np.transpose(arr) #transpose of a matrix
print("The transpose of the matrix is :")
print(t)
trace=np.trace(arr) #sum of diagonal elements
print("The of matrix is : ")
print(trace)
arr1=np.array([[8,3,1],
       [8,3,5],
       [3,6,4]])
print("The second array is : ")
print(arr1)
product=np.dot(arr,arr1) #matrix multiplication
print("The product of two matrix is")
print(product)
print("The inverse of the matrix one is :")
print(np.invert(arr)) #matrix inverse
```

```python
arr=np.array([[1,2,3],
      [4,5,6],
      [8,2,4]])
invers=np.linalg.inv(arr)
print(invers)
print("Rank of matrix is :",np.linalg.matrix_rank(arr))
print("Eigen value is : ")
print(np.linalg.eig(arr))
print("Eigen vector : ",np.linalg.eigvals)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3])
y=np.array([2,5,6])
plt.plot(x,y,marker='h',color='purple',ms=20,mec='green',mfc='yellow') # mfc=antha pointkulla filling color change panna,mec=marker edge color,ms=point size change pannalam,color=line color kudukalam,marker='o' means point round varum,h->hexagon,s->sqare,*->star,d->diamond etc..
plt.plot(x,y,'o:r') #marker line color and pattern
plt.show()
```
